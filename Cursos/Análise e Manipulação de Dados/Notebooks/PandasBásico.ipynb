# Pandas

Pandas é uma biblioteca **open-source** feita principalmente para trabalhar com dados relacionais ou rotulados de forma fácil e intuitiva. 

Ele fornece várias estruturas de dados e operações para manipular dados numéricos e séries temporais. 

Esta biblioteca é construída no topo da biblioteca NumPy. 

Pandas é rápido e tem alto desempenho e produtividade para os usuários.

## Características

- Rápido e eficiente para manipular e analisar dados
- Dados de diferentes objetos de arquivo podem ser carregados
- Fácil manuseio de dados ausentes (representados como NaN) em ponto flutuante, bem como dados de ponto não flutuante
- Mutabilidade de tamanho: colunas podem ser inseridas e excluídas do DataFrame e objetos de dimensão superior
- Mesclagem e junção de conjuntos de dados
- Remodelagem flexível e dinamização de conjuntos de dados
- Fornece funcionalidade de série temporal
- Poderoso agrupamento por funcionalidade para realizar operações *split-apply-combine* em conjuntos de dados

## Estruturas de Dados

Pandas geralmente fornecem duas estruturas de dados para manipular dados, são elas:

- Series
- DataFrame

### Series

Pandas Series é um array rotulado unidimensional capaz de conter dados de qualquer tipo (inteiro, string, float, objetos python, etc.). Os rótulos dos eixos são chamados coletivamente de índice. Pandas Series nada mais é que uma coluna em uma planilha do Excel. Os rótulos não precisam ser exclusivos, mas devem ser do tipo hashable. O objeto oferece suporte para indexação inteira e baseada em rótulo e fornece uma série de métodos para executar operações envolvendo o índice.

#### Criando uma Series 

No mundo real, uma Series Pandas será criada carregando os conjuntos de dados do armazenamento existente, o armazenamento pode ser banco de dados SQL, arquivo CSV e arquivo Excel. A Series Pandas pode ser criada a partir de listas, dicionário e de um valor escalar, etc.

```python
import pandas as pd 
  
# Criando uma Series
data = list(range(10,21))
series = pd.Series(data) 
print(series)

# Verificando o tipo
print(type(series))
```

### DataFrame

O Pandas DataFrame é uma estrutura de dados tabular bidimensional mutável, potencialmente heterogênea, com eixos rotulados (linhas e colunas). Um DataFrame é uma estrutura de dados bidimensional, ou seja, os dados são alinhados de forma tabular em linhas e colunas. O Pandas DataFrame consiste em três componentes principais: **dados**, **linhas** e **colunas**.

#### Criando um DataFrame

No mundo real, um Pandas DataFrame será criado carregando os conjuntos de dados do armazenamento existente, o armazenamento pode ser banco de dados SQL, arquivo CSV e arquivo Excel. O Pandas DataFrame pode ser criado a partir de listas, dicionário e de uma lista de dicionários, etc.

```python
nomes = ['João','José','Maria','Luiz','Paula']
nascimentos = [968, 155, 77, 578, 973]

bebês = list(zip(nomes,nascimentos))

df = pd.DataFrame(data=bebês, columns=['Nomes','Nascimentos'])
df
```

#### Convertendo o DataFrame para CSV

Uma vez que temos um objeto DataFrame, podemos facilmente transformá-lo CSV.

```python
print(df.to_csv())
```

#### Convertendo o DataFrame para JSON

Da mesma forma, podemos converter o nosso DataFrame para JSON.

```python
print(df.to_json(force_ascii=False))
```

### Filtrando Dados

Pandas nos disponibiliza diversos métodos que nos permitem filtrar nossos dados de acordo com determinadas condições.

```python
dados.where(dados['Attack'] > 100).tail(10)
```

Obtendo apenas os Pokémons com **'Attack'** maior que 80 e menor que 120.

```python
dados[(dados['Attack'] > 80) & (dados['Attack'] < 120)]
```

Obtendo apenas os Pokémons com **'HP'** maior que 100 e **'Speed'** maior ou igual a 130.

```python
dados[(dados['HP'] > 100) & (dados['Speed'] >= 130)]
```

Podemos usar o método **filter()** com uma expressão regular para selecionar apenas os Pokémons que começam com a letra **P**.

```python
dados.filter(regex='^P',axis=0).head(15)
```

### Ordenando os Valores

A função Pandas **sort_values()** classifica um DataFrame em ordem crescente ou decrescente da coluna passada.

```python
dados.sort_values("Name", axis = 0, ascending = True)
```

```python
dados.sort_values("Name", axis = 0, ascending = False)
```

### GroupBy

Groupby é um conceito bastante simples. Podemos criar um agrupamento de categorias e aplicar uma função às categorias.

```python
grupo = dados.groupby('Type 1')
```

O atributo **groups** nos traz os diferentes grupos juntamente com os seus integrantes.

```python
print(grupo.groups)
```

Agora imprimimos as primeiras entradas em todos os grupos formados.

```python
grupo.first()
```

Podemos utilizar a função **max()** para obter o Pokémon mais forte de cada grupo.

```python
grupo.max()
```

Podemos utilizar a função **min()** para obter o Pokémon mais fraco de cada grupo.

```python
grupo.min()
```

Podemos também obter a **média** de cada grupo.

```python
grupo.mean()
```
