# Visualização de Dados

![img](https://i.ibb.co/41FqQtK/Data-Visualizationheader.png)

A visualização de dados é um campo interdisciplinar que trata da representação gráfica de dados

É uma forma particularmente eficiente de comunicação quando os dados são numerosos como, por exemplo, uma série temporal. Do ponto de vista acadêmico, essa representação pode ser considerada como um mapeamento entre os dados originais (geralmente numéricos) e os elementos gráficos (por exemplo, linhas ou pontos em um gráfico).

O mapeamento determina como os atributos desses elementos variam de acordo com os dados. Sob essa luz, um gráfico de barras é um mapeamento do comprimento de uma barra para a magnitude de uma variável.

A visualização de dados tem suas raízes no campo da Estatística e, portanto, é geralmente considerada um ramo da Estatística Descritiva. No entanto, como tanto habilidades de design quanto habilidades estatísticas e de computação são necessárias para visualizar com eficácia, alguns autores argumentam que isso é uma arte e uma ciência.

Mais informações podem ser obtidas no **[Wikipedia](https://en.wikipedia.org/wiki/Data_visualization)**

## Matplotlib

**[Matplotlib](https://matplotlib.org/)** é uma biblioteca compreensiva para a criação de visualizações estáticas, animadas e interativas em Python.

Neste notebook, iremos abordar os fundamentos de matplotlib, incluindo tópicos como:

- Funções Básicas
- Objetos Figures
- Subplots
- Styling

Começaremos importando matplotlib para experimentar suas capacidades.

```python
import matplotlib.pyplot as plt
```

### Plotando um Gráfico

Primeiramente vamos criar um simples gráfico de uma equação linear básica.

Vamos também usar a biblioteca **numpy** para gerar dados.

```python
import numpy as np

x = np.arange(0,10)
y = 2 * x

print(x)
print(y)

plt.plot(x,y);
```

### Customizando um Gráfico

Podemos customizar o nosso gráfico com:

- Título
- Rótulo x
- Rótulo y
- Incluir limite para x
- Incluir limite para y
- Definir uma grid

Por fim, podemos salvá-lo como `.png` ou `.jpg` com a função **savefig()**.

```python
plt.title('Título do Gráfico')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(0,9)
plt.ylim(0,18)
plt.grid()
plt.plot(x,y);
plt.savefig('gráfico.png')
```

### Objeto Figure

A [API Orientada a Objetos](https://matplotlib.org/stable/api/index.html) de matplotlib faz uso do objeto **Figure**.

Nós então podemos adicionar eixos (**axes**) a este objeto Figure e então plotamos esses eixos.

Isso permite um controle robusto sob o gráfico.

Vamos agora implementar Figuras e Eixos na prática.

Começaremos gerando dados para o nosso gráfico.

```python
a = np.linspace(0,10,11)
b = a**4

print(a)
print(b)
```

```python
x = np.arange(0,10)
y = 2 * x

print(x)
print(y)
```

Agora usaremos o objeto **Figure** para construir o gráfico.

Também vamos adicionar eixos ao objeto Figure.

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y);
axes1.set_xlim(0,9)
axes1.set_ylim(0,18)
axes1.set_xlabel('Eixo X')
axes1.set_ylabel('Eixo Y')
axes1.set_title('Gráfico Linear')

axes2 = fig.add_axes([0.15,0.6,0.3,0.3])
axes2.plot(a,b);
axes2.set_xlim(0,10)
axes2.set_ylim(0,10_000)
axes2.set_xlabel('Eixo X')
axes2.set_ylabel('Eixo Y')
axes2.set_title('Gráfico Exponencial');
```

#### Parâmetros Figure

Para customizar o nosso objeto Figure, vamos experimentar com diferentes parâmetros.

Podemos por exemplo aumentar o tamanho (**figsize**) e qualidade (**dpi**) de nosso gráfico.

Além disso, também podemos exportar uma Figure.

```python
fig = plt.figure(figsize=(9,6), dpi=100)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
fig.savefig('figure.png', bbox_inches='tight')
```

### Subplots

Subplots nos permite adicionar um eixo à figura atual ou recuperar um eixo existente.

Novamente criaremos dois conjuntos de dados para o nosso gráfico.

```python
a = np.linspace(-10,10,50)
b = a ** 3
```

```python
x = np.arange(-10,11)
y = abs(x)
```

Subplots nos permite definir o número de linhas e colunas que teremos.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

Vamos agora criar um gráfico com os nossos dados.

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,8))

axes[0].plot(x,y);
axes[0].set_xlabel('Eixo X',fontsize=18)
axes[0].set_ylabel('Eixo Y',fontsize=18)
axes[0].set_title('Gráfico 1',fontsize=20)
axes[0].grid()

axes[1].plot(a,b);
axes[1].set_xlabel('Eixo X',fontsize=18)
axes[1].set_ylabel('Eixo Y',fontsize=18)
axes[1].set_title('Gráfico 2',fontsize=20)
axes[1].grid()

st = fig.suptitle('Título Principal', fontsize=24)
st.set_y(1.025)
fig.tight_layout()
fig.savefig('subplots.png', bbox_inches='tight')
```

A seguir vamos plotar dois gráficos: **cosseno** e **seno**.

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14,8))

x = np.linspace(0,11,100)
y = np.cos(x)
z = np.sin(x)

axes[0].plot(x,y, color='red', lw=2.5)
axes[0].set_title('Cosseno', fontsize=20)
axes[0].set_xlim(0,11)
axes[0].set_xlabel('X', fontsize=15, c='r')
axes[0].set_ylabel('Cosseno X', fontsize=15, c='r')
axes[0].grid(linestyle='--')

axes[1].plot(x,z, color='green', lw=2.5);
axes[1].set_title('Seno', fontsize=20)
axes[1].set_xlim(0,11);
axes[1].set_xlabel('X', fontsize=15, c='g')
axes[1].set_ylabel('Seno X', fontsize=15, c='g')
axes[1].grid(linestyle=':')

fig.subplots_adjust(hspace=0.35)
```

### Legendas

Vamos aprender a adicionar legendas aos nossos gráficos.

Novamente usaremos um objeto **Figure**.

```python
fig = plt.figure(figsize=(9,6))

x = np.linspace(0,11,20)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x*8, label='X vs X*10')
ax.plot(x,x**2, label='X vs X^2')
ax.grid()
ax.legend(loc='lower right');
```

### Cores e Estilos

Nosso foco agora será em estilos visuais de nosso gráfico.

Estilos possíveis de linha: `--`, `-`, `-.`, `:`, `steps`

Cores podem ser definidas por nomes ou código hexadecimal.

Vamos então definir mais um gráfico.

```python
z = np.linspace(-11,11,30)

fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0,0,1,1])
ax.grid()
ax.plot(z, z*z, color='#069139', linewidth=3, label='z vs $z^2$', ls='--', marker='s', ms=10);
ax.plot(z, z*z+30, color='purple', lw=3, label='z vs $z^z+30$', linestyle='-.', marker='o', markersize=10);
ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_color('blue')
ax.spines['bottom'].set_linewidth(3)
ax.legend(loc='upper center', prop={'size':15});
```

## Seaborn

[Seaborn](https://seaborn.pydata.org/index.html) é uma biblioteca de visualização de dados Python baseada em matplotlib. Ele fornece uma interface de alto nível para desenhar gráficos estatísticos atraentes e informativos.

Seaborn também é capaz de interagir perfeitamente com os **DataFrames** da biblioteca pandas.

### Scatter Plot

Um gráfico de dispersão é um tipo de gráfico ou diagrama matemático que usa coordenadas cartesianas para exibir valores para normalmente duas variáveis de um conjunto de dados. Os dados são exibidos como uma coleção de pontos, cada um tendo o valor de uma variável determinando a posição no eixo horizontal e o valor da outra variável determinando a posição no eixo vertical.

Vamos explorar **scatter plots** com seaborn, para isso, vamos usar dados de vendas de itens tecnológicos, por isso também usaremos a biblioteca pandas.

```python
import pandas as pd
import seaborn as sns
```

Carregamos os dados em um **DataFrame**.

```python
df = pd.read_csv('dados_seaborn/office_sales.csv')
df.head()
```

Podemos visualizar a relação entre **salário** e **vendas** coloridos por **nível de educação** com um **scatter plot**.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='salary', y='sales', hue='level of education', alpha=0.85);
```

### Distribution Plot

Um gráfico de distribuição exibe uma distribuição e intervalo de um conjunto de valores numéricos plotados em relação a uma dimensão.

```python
df['division'].value_counts()
```

Aqui podemos ver um **countplot** por divisão.

```python
plt.figure(figsize=(10,6), dpi=120)
sns.countplot(data=df, x='division');
```

Aqui podemos ver um **countplot** por nível de educação.

```python
plt.figure(figsize=(10,6), dpi=120)
sns.countplot(data=df, x='level of education');
```

Aqui vemos um gráfico de barras por nível de educação e salário médio (mean).

**sd** indica standard deviation (desvio padrão) e é apresentado como uma reta cinza no gráfico.

```python
plt.figure(figsize=(10,6), dpi=120)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd');
```

No gráfico a seguir incluímos também a divisão do trabalhador (em diferentes cores), além do seu nível de educação e salário.

```python
plt.figure(figsize=(10,6), dpi=120)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division');
plt.legend(bbox_to_anchor=(1.05,1));
```

Para visualizarmos apenas a distribuição dos salários podemos usar um **histplot**.

```python
plt.figure(figsize=(10,6), dpi=120)
sns.histplot(df['salary']);
```

```python
df['salary'].max()
```

### Dados Estudantes

Agora vamos carregar um novo conjunto de dados para experimentarmos outros tipos de gráficos disponíveis em **seaborn**.

Trata-se de dados sobre a perfomance de estudantes em diferentes testes.

```python
data = pd.read_csv('dados_seaborn/StudentsPerformance.csv')
data.head()
```

### Box Plot

Na estatística descritiva, um box plot é um método para representar graficamente grupos de dados numéricos por meio de seus quartis. Os gráficos de caixa também podem ter linhas que se estendem das caixas, indicando variabilidade fora dos quartis superior e inferior.

Vamos ver a relação de distribuição de pontuação em matemática para estudantes que fizeram curso preparatório e estudantes que não fizeram.

```python
plt.figure(figsize=(9,5), dpi=120)
sns.boxplot(data=data, y='math score', x='test preparation course');
```

Também podemos ver do teste de leitura.

```python
plt.figure(figsize=(9,5), dpi=120)
sns.boxplot(data=data, y='reading score', x='test preparation course');
```

Também podemos ver o quanto que o nível de educação dos pais pode afetar a pontuação no teste de leitura.

```python
plt.figure(figsize=(12,5), dpi=120)
sns.boxplot(data=data, y='reading score', x='parental level of education');
```

Podemos ainda incluir se o estudante completou ou não o curso preparatório.

```python
plt.figure(figsize=(12,5), dpi=120)
sns.boxplot(data=data, y='reading score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1,0.7));
```

Também podemos inverter a orientação de nosso gráfico.

```python
plt.figure(figsize=(12,5), dpi=120)
sns.boxplot(data=data, x='reading score', y='parental level of education', hue='test preparation course', palette='Set2')
plt.legend(bbox_to_anchor=(1,0.7));
```

### Violin Plot

Uma trama de violino é um método de traçar dados numéricos. É semelhante a um gráfico de caixa, com a adição de um gráfico de densidade do kernel girado em cada lado.

Os gráficos de violino são semelhantes aos gráficos de caixa, exceto que também mostram a densidade de probabilidade dos dados em valores diferentes, geralmente suavizados por um estimador de densidade de kernel. Normalmente, um gráfico de violino incluirá todos os dados que estão em um gráfico de caixa: um marcador para a mediana dos dados; uma caixa ou marcador indicando o intervalo interquartil; e possivelmente todos os pontos de amostra, se o número de amostras não for muito alto.

Vejamos como um gráfico de violino funciona.

```python
plt.figure(figsize=(12,8), dpi=120)
sns.violinplot(data=data, x='reading score', y='parental level of education');
```

### Comparison Plot

O gráfico de comparação permite que o usuário compare uma determinada estatística entre conjuntos de dados e populações.

Vejamos um exemplo de um **jointplot** comparando a pontuação de estudantes em matemática vs leitura.

```python
plt.figure(figsize=(9,9), dpi=120)
sns.jointplot(data=data, x='math score', y='reading score');
```

Também podemos mudar o tipo de gráfico para 'hex', o padrão é **scatter**.

```python
plt.figure(figsize=(9,9), dpi=120)
sns.jointplot(data=data, x='math score', y='reading score', kind='hex');
```

Outra opção que temos é alterar o **hue** que irá separar nossos pontos em cores diferentes de acordo com a categoria.

```python
plt.figure(figsize=(9,9), dpi=120)
sns.jointplot(data=data, x='math score', y='reading score', hue='gender');
```

O **pairplot** plota relacionamentos de pares em um conjunto de dados.

```python
sns.pairplot(data=data, hue='gender');
```

Para evitar os gráficos duplicados, podemos setar `corner=True`.

```python
sns.pairplot(data=data, hue='gender', corner=True);
```

### Heatmap

Um Heatmap (ou mapa de calor) é uma técnica de visualização de dados que mostra a magnitude de um fenômeno como cor em duas dimensões. A variação na cor pode ser por **hue** ou intensidade, dando pistas visuais óbvias ao leitor sobre como o fenômeno está agrupado ou varia no espaço.

Para compreendermos melhor os Heatmaps, vamos carregar um novo conjunto de dados com diferentes informações sobre países.

```python
paises = pd.read_csv('dados_seaborn/country_table.csv')
paises.head(8)
```

Vamos agora definir os países como índice de nosso **DataFrame**.

```python
paises = paises.set_index('Countries')
paises.head()
```

Podemos então facilmente apresentar um **heatmap**.

Removeremos a coluna **Life expectancy**, pois está em uma unidade diferente.

```python
plt.figure(dpi=110)
sns.heatmap(paises.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis');
```

Temos também a opção de usar um **clustermap**, que é semelhante ao heapmap, porém agrupa valores semelhantes.

```python
plt.figure(dpi=110)
sns.clustermap(paises.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis');
```
