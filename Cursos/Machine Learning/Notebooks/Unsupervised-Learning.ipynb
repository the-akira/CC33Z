# Unsupervised Learning

**Unsupervised Learning** (UL) é um tipo de algoritmo que aprende padrões a partir de dados não rotulados. 

Dois dos principais métodos usados no unsupervised learning são a análise de **principal component** e **cluster**.

No unsupervised learning, os algoritmos são deixados sozinhos para descobrir estruturas interessantes nos dados.

Os dados fornecidos aos **algoritmos unsuperviseds** não são rotulados, o que significa que apenas as variáveis de **input** (x) são fornecidas sem variáveis de **output** correspondentes. No unsupervised learning, os algoritmos são deixados para descobrir estruturas interessantes nos dados por conta própria.

No supervised learning, o sistema tenta aprender com os exemplos anteriores dados. No unsupervised learning, o sistema tenta encontrar os padrões diretamente do exemplo dado. Portanto, se o conjunto de dados estiver rotulado, é um problema **supervised**, e se o conjunto de dados não estiver rotulado, é um problema **unsupervised**.

**Terminologias Importantes**

- **Feature**: Uma variável de input usada para fazer previsões.

- **Predictions**: O output de um modelo quando fornecido com um exemplo de input.

- **Example**: Uma linha de um conjunto de dados. Um exemplo contém um ou mais features e possivelmente um label.

- **Label**: Resultado da feature.

## Clusterização

Na clusterização, os dados são divididos em vários grupos com características semelhantes.

O [K-Means Clustering](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html) é um algoritmo de clusterização iterativo que visa encontrar máximos locais em cada iteração. Inicialmente, o número desejado de clusters é escolhido. Em nosso exemplo, sabemos que há três classes envolvidas, então programamos o algoritmo para agrupar os dados em três classes, passando o parâmetro **“n_clusters”** para nosso modelo K-Means. Aleatoriamente, três pontos (inputs) são atribuídos em três grupos. Com base na distância do centróide entre cada ponto, os próximos inputs fornecidos são segregados em clusters respeitados e os centróides são recalculados para todos os clusters.

Cada centróide de um cluster é uma coleção de valores de features que definem os grupos resultantes. O exame dos pesos dos features do centróide pode ser usado para interpretar qualitativamente que tipo de grupo cada cluster representa.

Importamos então o modelo K-Means da biblioteca do Scikit-Learn, ajustamos os features e fazemos previsões.

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)

model.fit(iris_data.data)

predicted_label = model.predict([[7.2, 3.5, 0.8, 1.6]])
all_predictions = model.predict(iris_data.data)
```

Imprimimos as nossas previsões.

```python
print(predicted_label)
print(all_predictions)
```

Comparamos com a verdade absoluta.

```python
iris_data.target
```

Avaliamos o desempenho de nosso modelo através da accuracy.

```python
correct_pred = []

for i,pred in enumerate(all_predictions):
    if pred == iris_data.target[i]:
        correct_pred.append(pred)
```

```python
acc = len(correct_pred) / len(all_predictions)
acc
```

Como podemos observar, nosso algoritmo não teve um desempenho adequado para este conjunto de dados.
