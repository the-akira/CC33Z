# Gráficos SymPy

SymPy usa a biblioteca Matplotlib como backend para renderizar gráficos **2-D** e **3-D** de funções matemáticas.

O suporte à plotagem é definido no módulo **sympy.plotting**. 

As seguintes funções estão presentes no módulo de plotagem:

- **plot** - plotagens de linha 2D

- **plot3d** - gráficos de linha 3D

- **plot_parametric** - plotagens paramétricas 2D

- **plot3d_parametric** - plotagens paramétricas 3D

A função **plot()** retorna uma instância da classe **Plot**. 

Uma figura do gráfico pode ter uma ou mais expressões SymPy.

```
plot(expr, range, kwargs)
```

Onde **expr** é qualquer expressão SymPy válida. Se não for mencionado, o intervalo (**range**) usa o padrão como (-10, 10).

O exemplo a seguir plota valores de $2x + 3$ para cada valor no intervalo (-10,10):

````python
from sympy import *

x = Symbol('x') 
plot(2*x + 3,line_color='red');
````

A seguir vamos desenhar o gráfico da função: $f(x) = |x|$

````python
plot(Abs(x));
````

A seguir vamos desenhar o gráfico da função: $f(x) = x^2 - 1$

````python
plot(x**2 - 1, (x, -2, +2), line_color='black');
````

A seguir vamos desenhar o gráfico da função: $f(x) = x^3$

````python
plot(x**3,line_color='purple');
````

Para desenhar vários gráficos para o mesmo intervalo, forneça várias expressões antes da tupla de intervalo:

````python
plot(sin(x),cos(x),(x, -pi, pi));
````

Podemos também especificar um intervalo separado para cada expressão:

```
plot((expr1, range1), (expr2, range2))
```

A figura a seguir plota **sin(x)** e **cos(x)** em intervalos diferentes:

````python
plot((sin(x),(x,-2*pi,2*pi)), (cos(x),(x,-pi,pi)));
````

Os seguintes argumentos opcionais de palavra-chave podem ser especificados na função **plot()**:

- **line_color** - Especifica a cor da linha do gráfico.
- **title** - Uma string a ser exibida como título
- **xlabel** - Uma string a ser exibida como rótulo para o eixo X
- **ylabel** - Uma string a ser exibida como rótulo para o eixo y

````python
plot((sin(x),(x,-2*pi,2*pi)), (cos(x),(x,-pi,pi)), line_color='green', title='Gráfico SymPy');
````

Podemos também alterar a cor apenas de uma série específica:

````python
expr1 = (x, cos(2*pi*x)/2 + 1/2)
expr2 = (x, sin(2*pi*x)/2 + 1/2)
p = plot_parametric(expr1, expr2, (x, 0, 1))
````

````python
p[0].line_color = 'red';
p.show()
````

Com este conhecimento, podemos então customizar nossos gráficos.

Por exemplo:

````python
p = plot(x**2, (x,-3,+3), legend=True, line_color='r', show=False)
p.extend(plot((x**2-2), (x,-3,+3), legend=True, line_color='b', show=False))
p.extend(plot((x**2+2), (x,-3,+3), legend=True, line_color='g', show=False))
p.show()
````

A função **plot3d()** renderiza um gráfico tridimensional:

```
plot3d(expr, xrange, yrange, kwargs)
```

O exemplo a seguir desenha um gráfico de superfície 3D:

````python
from sympy.plotting import plot3d 
x,y = symbols('x y') 
plot3d(x*y, (x,-10,10), (y,-10,10));
````

Como no gráfico 2D, um gráfico tridimensional também pode ter vários gráficos, cada um com um intervalo diferente:

````python
plot3d(x*y, x/y, (x,-5,5), (y,-5,5));
````

A função **plot3d_parametric_line()** renderiza um gráfico de linha paramétrica tridimensional:

````python
from sympy.plotting import plot3d_parametric_line 
plot3d_parametric_line(cos(x), sin(x), x, (x,-5,5));
````

Para desenhar um gráfico de superfície paramétrico, usamos a função **plot3d_parametric_surface()**:

````python
from sympy.plotting import plot3d_parametric_surface 
plot3d_parametric_surface(cos(x+y), sin(x-y), x-y, (x,-5,5), (y,-5,5));
````
