```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
plt.style.use('classic')
```

```python
def derivada(f,x):
    """
    Função que recebe uma função e um ponto 
    e calcula a determinada tangente
    """
    h = 0.000000001 # tamanho dos passos
    return (f(x+h) - f(x)) / h # definição da derivada
```

```python
def linha_tangente(f,x_0,a,b):
    x = np.linspace(a,b,100)
    y = f(x) 
    y_0 = f(x_0)
    y_tan = derivada(f,x_0) * (x - x_0) + y_0 
    
    plt.figure(figsize=(10,7))
    plt.grid()
    plt.plot(x,y,'r-',linewidth=2)
    plt.plot(x,y_tan,'b-',linewidth=2)
    plt.plot(x_0,y_0,'k',marker='o')
    plt.xlim([a,b])
    plt.ylim(-1,6)
    plt.xlabel('x')     
    plt.ylabel('y')    
    plt.title('Função com a Linha Tangente') 
    plt.show()    
```

```python
def f1(x):
    return x**2

linha_tangente(f1,-0.5,-3,3)
```

```python
x = sym.Symbol('x')
sym.diff(x**2,x)
```

```python
def df1(x):
    return 2 * x
```

```python
x = np.linspace(-3,3,200)
plt.title('Função derivada')
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.plot(x, df1(x), c='r', linewidth=2)
plt.plot(-0.5,-1,'k',marker='o')
plt.axhline(y=-1,xmax=0.25,color='k',linestyle='dashed')
plt.axvline(x=-0.5,ymax=0.25,color='k',linestyle='dashed')
plt.axis([-1,1,-2,2])
plt.grid()
```
